```python
def convertir_columnas_a_float(df, columnas):
    'convierte las columnas object en float'
    for col in columnas:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].astype(float)
    return df
df = convertir_columnas_a_float(df, ['O2', 'CO', 'CO2', 'T_air', 'CO (11.0%)'])
```

```python
# Crear DataFrame nuevo vac√≠o con las mismas columnas del original

df_nuevo = pd.DataFrame(columns=df.columns)

# Definir funci√≥n para aplicar transformaciones
def transformar_columnas(df_grupo):
    df_grupo['Date'] = df['Date']
    df_grupo['Time'] = df['Time']
    df_grupo['O2'] = round(df_grupo['O2'] + np.random.uniform(0.5, 1.5), 1)
    df_grupo['CO'] = round(df_grupo['CO'] + np.random.uniform(0.5, 1.5),0)
    df_grupo['CO2'] = round((-0.55629 * df_grupo['O2'] + 11.7435), 1)
    df_grupo['T_air'] = round(df_grupo['T_air'] + np.random.uniform(0.5, 1.5), 1)
    df_grupo["CO (11.0%)"] = round(df_grupo["CO"]*(10/(21-df_grupo["O2"])), 0)
    
    nuevo_df = pd.DataFrame({
        'Date': df_grupo['Date'],
        'Time': df_grupo['Time'],
        'O2': df_grupo['O2'],
        'CO': df_grupo['CO'],
        'CO2': df_grupo['CO2'],
        'T_air': df_grupo['T_air'],
        'CO (11.0%)': df_grupo["CO (11.0%)"]
    })

    return nuevo_df
```

```python
#Crear nuevo DataFrame con las transformaciones
nuevo_df= transformar_columnas(df)
```

```python
#Concatenar df y nuevo_df
gas = pd.concat([df, nuevo_df], axis=0)
```

```python
#Convertir los archivos .csv en DataFrame y concatenar

frame_T = pd.read_csv(datos_path_T + '\\' + files_T[0], sep = '\t', decimal=",", header=None)

for f in files_T[1:]:    
    frame_T = frame_T.append(pd.read_csv(datos_path_T + '\\' + f, sep = '\t', decimal=",", header=None))
    
df_T = frame_T.drop([0, 1, 5], axis = 1)
df_T.rename({2: 'PC', 3: 'C', 4: 'CH'}, axis=1)
df_T.columns = ['PC', 'C', 'CH']
```

```python
def convertir_columnas_a_float(df, columnas):
    'convierte las columnas object en float'
    for col in columnas:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].astype(float)
    return df
df_T = convertir_columnas_a_float(df_T, ['PC', 'C', 'CH'])
```

```python
# Crear DataFrame nuevo vac√≠o con las mismas columnas del original
df_T_nuevo = pd.DataFrame(columns=df_T.columns)

# Definir funci√≥n para aplicar transformaciones
def transformar_columnas(df_grupo):
    df_grupo['PC'] = round(df_grupo['PC'] + np.random.uniform(0.6, 1.6), 1)
    df_grupo['C'] = round(df_grupo['C'] + np.random.uniform(0.6, 1.6), 1)
    df_grupo['CH'] = round(df_grupo['CH'] + np.random.uniform(0.6, 1.6), 1)
      
    nuevo_T_df = pd.DataFrame({
        'PC': df_grupo['PC'],
        'C': df_grupo['C'],
        'CH': df_grupo['CH'],

    })

    return nuevo_T_df
```

```python
#Crear nuevo DataFrame con las transformaciones
nuevo_T_df= transformar_columnas(df_T)
```

```python
nuevo_T_df.shape
```

```python
#Eliminar los datos nulos
nuevo_T_df.dropna(inplace=True)
```

```python
nuevo_T_df.shape
```

```python
nuevo_T_df.head(10)
```

```python
#Seleccionar valores que cumplen con la normativa
New_temp = nuevo_T_df[(nuevo_T_df['PC']>=890)&(nuevo_T_df['C']>721)&(nuevo_T_df['CH']<=250)]
```

```python
# Crear una lista de dataframes con cada grupo de 10 filas y su promedio
filas = []
for i in range(0, len(New_temp), 10):
    grupo = New_temp.iloc[i:i+10]
    promedio = round(grupo.mean(), 1)
    promedio_df = pd.DataFrame(promedio).transpose()
    filas.append(grupo)
    filas.append(promedio_df)

# Concatenar los dataframes en uno solo
temp = pd.concat(filas).reset_index(drop=True)
```
