```python
def convertir_columnas_a_float(df, columnas):
    'convierte las columnas object en float'
    for col in columnas:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].astype(float)
    return df
df = convertir_columnas_a_float(df, ['O2', 'CO', 'CO2', 'T_air', 'CO (11.0%)'])
```

```python
# Crear DataFrame nuevo vacío con las mismas columnas del original

df_nuevo = pd.DataFrame(columns=df.columns)

# Definir función para aplicar transformaciones
def transformar_columnas(df_grupo):
    df_grupo['Date'] = df['Date']
    df_grupo['Time'] = df['Time']
    df_grupo['O2'] = round(df_grupo['O2'] + np.random.uniform(0.5, 1.5), 1)
    df_grupo['CO'] = round(df_grupo['CO'] + np.random.uniform(0.5, 1.5),0)
    df_grupo['CO2'] = round((-0.55629 * df_grupo['O2'] + 11.7435), 1)
    df_grupo['T_air'] = round(df_grupo['T_air'] + np.random.uniform(0.5, 1.5), 1)
    df_grupo["CO (11.0%)"] = round(df_grupo["CO"]*(10/(21-df_grupo["O2"])), 0)
    
    nuevo_df = pd.DataFrame({
        'Date': df_grupo['Date'],
        'Time': df_grupo['Time'],
        'O2': df_grupo['O2'],
        'CO': df_grupo['CO'],
        'CO2': df_grupo['CO2'],
        'T_air': df_grupo['T_air'],
        'CO (11.0%)': df_grupo["CO (11.0%)"]
    })

    return nuevo_df
```

```python
#Crear nuevo DataFrame con las transformaciones
nuevo_df= transformar_columnas(df)
```

```python
# resetear los índices de ambos dataframes
gas = gas.reset_index(drop=True)
temp = temp.reset_index(drop=True)

# unir los dataframes horizontalmente
df_concat = pd.concat([gas, temp], axis=1)
```

```python
df_concat.head(12)
```

```python
#Agregar las columnas de %EC y %EA

df_concat['%EC'] = round(df_concat.apply(lambda row: 100-(0.39*((row['CH']- row['T_air'])/row['CO2'])), axis=1), 1)
df_concat['%EA'] = round(df_concat.apply(lambda row: 1.025*row['CO2']**2 - 32.012*row['CO2'] + 235.29, axis=1), 1)
```

```python
df_concat.head(11)
```

```python
# Convertir la columna de fechas a formato datetime
df_concat['Date'] = pd.to_datetime(df_concat['Date'], format='%m/%d/%Y')
```

```python
# Modificar fechas
df_concat['Date'] = df_concat['Date'].apply(lambda x: x.replace(day=3, month=3, year=2023))
```

```python
df_concat
```

```python
# Convertir la columna de horas a formato datetime
df_concat['Time'] = pd.to_datetime(df_concat['Time'])
```

```python
# convierte la columna 'Time' a formato de 24 horas
df_concat['Time'] = df_concat['Time'].apply(lambda x: x.strftime('%H:%M:%S'))
```

```python
# elimina los últimos dos caracteres que representan AM/PM
df_concat['Time'] = df_concat['Time'].apply(lambda x: x[:-3])
```

```python
# define una función que agregue horas y minutos a cada tiempo
def agregar_tiempo(time):
    time = datetime.strptime(time, '%H:%M')
    time = time + timedelta(hours=2, minutes=30)
    return time.strftime('%H:%M')
```

```python
# aplica la función a las filas seleccionadas de la columna 'Time'
df_concat.loc[1:2, 'Time'] = df_concat.loc[1:2, 'Time'].apply(agregar_tiempo)
```
